Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

# Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
def f_p():
    number = input("Enter document number: ")
    if number:
        for i in documents:
            if i['number'] == number:
                print('Name: ', i['name'])
                return
    print("Can't find document number")
    
    
def f_s():
    number = input("Enter document number: ")
    if number:
        for i in directories:
            if directories[i].count(number) > 0:
                print('Document is in directory #', i)
                return
    print("Can't find document number")
    
    
def f_l():
    for i in documents:
        print(i['type'], '"', i['number'], '"', '"', i['name'], '"')
    
    
def f_as():
    new_dir = input('Enter new directory number: ')
    
    if new_dir:
        if new_dir not in directories:
            directories[new_dir] = []
            print("Done!")
        else:
            print('We already have such directory name')
    else:
        print('Please enter valid directory number')

    
    
def f_d():
    number = input("Enter document number: ")
    if number:
        for i in documents:
            if i['number'] == number:
                documents.remove(i)
                
        for i in directories:
            if directories[i].count(number) > 0:
                directories[i].remove(number)
                print("Done!")
                return

    print("Can't find document number")
    
def f_m():
    number = input("Enter document number: ")
    new_dir = input('Enter new directory number: ')

    if number and new_dir in directories:
        for i in directories:
            if directories[i].count(number) > 0:
                directories[i].remove(number)
                directories[new_dir].append(number)
                print("Done!")
                return
    else:
        print("Can't find directory")
                
    print("Can't find document number")
    
    
def f_a():
    doc_type = input('Enter document type: ')
    number = input("Enter document number: ")
    name = input("Enter document's holder name: ")
    dir_num = input('Enter directory number: ')
    
    if dir_num not in directories:
        print('Please enter valid directory name')
        return
    
    if doc_type and number and name:
        new_doc = {'type': doc_type, 'number': number, 'name': name}
        documents.append(new_doc)
        directories[dir_num].append(number)
        print("Done!")
    else:
        print('Please enter valid params')

def main():
    command = input('Enter command code: ')
    if command == 'p':
        f_p()
    elif command == 's':
        f_s()
    elif command == 'l':
        f_l()
    elif command == 'as':
        f_as()
    elif command == 'd':
        f_d()
    elif command == 'm':
        f_m()
    elif command == 'a':
        f_a()
    else:
        print("Can't find command code. Please try enter valid command again")
        
while True:
    main()
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.
